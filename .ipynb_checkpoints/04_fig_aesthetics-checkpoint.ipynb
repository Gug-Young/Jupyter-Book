# Use Case: Figure Aesthetics (Legend/Colorbar Tips)
연구실에서 많이 겪는 **범례 위치, 컬러바 라벨 위치, 빈 축(divider) 조정** 문제를 최소 예제로 정리합니다.


```python

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable

# Sample data
x = np.linspace(-3, 3, 200)
y = np.linspace(-3, 3, 200)
X, Y = np.meshgrid(x, y)
Z = np.exp(-(X**2 + Y**2))

fig, ax = plt.subplots(dpi=140)
im = ax.imshow(Z, extent=[x.min(), x.max(), y.min(), y.max()], origin="lower")
ax.set_title("Legend & Colorbar Tweaks")

# 1) Legend 조금 더 오른쪽으로: bbox_to_anchor 사용
l0, = ax.plot(x, np.sin(x), label="sin(x)", lw=1.5, alpha=0.7)
l1, = ax.plot(x, np.cos(x), label="cos(x)", lw=1.5, alpha=0.7)
# loc과 bbox_to_anchor의 조합으로 미세 조정
leg = ax.legend(loc="upper right", bbox_to_anchor=(1.12, 1.02), frameon=False, ncol=1, fontsize=9)

# 2) Colorbar 라벨의 '위치'를 살짝 위로 - labelpad 조정
divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="4%", pad=0.05)
cbar = plt.colorbar(im, cax=cax)
cbar.set_label("P_ij", labelpad=8)  # labelpad로 위/아래 간격 조절
# outline 색 바꾸기
cbar.outline.set_edgecolor("gray")

# 3) '빈 박스'를 길고 아래로 내리기: axes_grid1로 별도 축 추가 후 스팬 조정 예시
#    (실데이터 대신 안내선으로 형태 확인)
cax2 = divider.append_axes("right", size="6%", pad=0.20)   # pad를 늘리면 '더 아래'로 내려온 느낌
cax2.set_ylim(0, 1)
cax2.set_xlim(0, 1)
cax2.spines[:].set_visible(True)
cax2.set_xticks([]); cax2.set_yticks([])
cax2.set_title("빈 박스", fontsize=9)
# 상자 모양 가이드
cax2.plot([0,1,1,0,0],[0,0,1,1,0], lw=1)

plt.tight_layout()
plt.show()
```

### 추가 팁
- `bbox_to_anchor=(x, y)`에서 `x>1.0`이면 그림 바깥쪽으로 우측 이동합니다.
- `colorbar.set_label(..., labelpad=값)`으로 라벨 위치를 미세 조정합니다.
- **axes_grid1**의 `append_axes`로 보조 축을 만들어 **길이/패딩**을 조절할 수 있습니다.
- `PathCollection`의 산점도 빈 마커는 `facecolors='none'`로 지정하세요 (`fillstyle`은 라인 스타일에만 적용되는 경우가 있어 에러가 날 수 있습니다).
