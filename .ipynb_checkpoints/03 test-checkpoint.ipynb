```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0,2*np.pi)
y = np.cos(x)
```

```python
a = [1, 2, 3, 4, 5]
sum(a)
```

:::{note} 실행 결과 확인
{eval}`a`

해당 부분은 `{eval}'sum(a)'`를 한 결과이다. 

{eval}`sum(a)`

글자와 글자 사이에 들어가게 되며 자연스럽게 보인다.
:::

## Figure eval 삽입

`{eval}`을 사용하면 글자사이에 그림도 넣을 수 있다.

```python
fig,ax = plt.subplots(1,1,figsize=(1.1,0.5))
plt.plot(x,y)
```

이 그림은 $y=sin(x)$를 통해서 얻어진 그림 {eval}`fig`이다.
