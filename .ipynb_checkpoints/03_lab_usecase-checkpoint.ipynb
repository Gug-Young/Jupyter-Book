# Use Case: Our Lab Workflow as a Narrative

연구실에서 자주 쓰는 **시뮬레이션 결과 정리 → 그림 정리 → 수식 요약**을 하나의 문서처럼 구성해봅니다.

## 1. 핵심 아이디어 요약
:::{admonition} TL;DR
:class: tip
- 실험/시뮬레이션마다 *반복되는 보고서 포맷*을 Jupyter Book으로 표준화합니다.
- 코드 셀은 **재현성**, 마크다운 셀은 **서술/결론**, 수식은 **정확성**을 담당합니다.
:::

## 2. 수식 블록 (LaTeX)
Kuramoto 계의 질량항이 포함된 변형에서 (illustrative),
$$ \dot{\theta}_i = \omega_i + K r \sin(\psi - \theta_i) $$
여기서 $r e^{i\psi} = \frac{1}{N}\sum_j e^{i\theta_j}$ 입니다.
