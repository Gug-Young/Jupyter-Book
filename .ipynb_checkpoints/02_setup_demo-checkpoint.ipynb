# Use Case: Minimal End-to-End Demo

이 노트북은 **데이터 생성 → 분석 → 시각화 → 간단한 결론**까지 한 번에 보여주는 최소 예시입니다.

## 목표
- 간단한 Kuramoto-like 토이모델에서 위상 동기화 지표 `r(t)`를 계산해봅니다.
- 그래프를 출력하고, 결과를 한두 문장으로 정리합니다.

:::{note}
이 노트북은 *발표 시연용*입니다. 실행 결과를 미리 저장해둔 뒤 Jupyter Book으로 빌드하면, 웹에서 그대로 보입니다.
:::


```python

import numpy as np
import matplotlib.pyplot as plt

# --- Toy Kuramoto-like setup (small N for demo) ---
np.random.seed(0)
N = 200
K = 1.5               # coupling
dt = 0.01
T = 20.0
steps = int(T/dt)
omega = np.random.normal(0.0, 1.0, N)
theta = np.random.uniform(-np.pi, np.pi, N)

r_hist = []
for _ in range(steps):
    # Order parameter
    re = np.mean(np.cos(theta))
    im = np.mean(np.sin(theta))
    r = np.sqrt(re**2 + im**2)
    r_hist.append(r)
    psi = np.arctan2(im, re)

    # Kuramoto update (Euler)
    theta += (omega + K * r * np.sin(psi - theta)) * dt

# --- Plot ---
plt.figure(dpi=120)
plt.plot(np.linspace(0, T, steps), r_hist, lw=2)
plt.xlabel("t")
plt.ylabel("r(t)")
plt.title("Toy Kuramoto-like synchronization index r(t)")
plt.grid(True)
plt.show()
```

## 간단 결론
- 초기 분산이 큰 자연주파수 분포에서도, **충분히 큰 결합 강도 $K$**에서는 `r(t)`가 증가하며 부분 동기화가 관찰됩니다.
